# Reinforcement Learning (RL)

## 1. Introduction

Reinforcement Learning (RL) is a machine learning paradigm in which an agent learns to make decisions in an uncertain and dynamic environment through experience. The agent's goal is to maximize a numerical reward it receives from the environment based on its actions.

In RL, the agent interacts with the environment through a series of states and actions. The agent observes the current state of the environment, takes an action, and receives a reward from the environment. The agent's objective is to find the optimal policy, i.e., the best way to act in each state to maximize the accumulated long-term reward.

To better understand Reinforcement Learning, it is important to understand its basic components:
 * <b>Agent</b>: The entity that makes decisions in the environment. Its objective is to get the maximum reward.
 * <b>Environment</b>: Represents the "world" where the actions are taken over an state and which the agent interacts.
 * <b>Interpreter</b>: The component that looks at the environment and informs the agent about the new state and the reward obtained after taking an action. 
 

<I><small>The typical framing of a Reinforcement Learning (RL) scenario: an agent takes actions in an environment, which is interpreted into a reward and a representation of the state, which are feedback into the agent (Source: https://en.wikipedia.org/wiki/Reinforcement_learning).</small></I>

We have other important RL concepts that must be knowed:
 * <b>Reward</b>: is a numerical signal that tells the agent how well it is performing in the environment.
 * <b>Policy</b>: is a function that determines the probability of the agent taking a particular action in a given state.
 * <b>Value function</b>: is a function that assigns a numerical value to each state or state-action pair representing the expected long-term reward.

RL is a good paradigm to find a good policy in episodic problems. That means that the problem can be defined as a problem with some end-state and it can be played/simulated multiple times. This is because the agent will learn from the experience.

Once we know the principal components, it is important to know the different phases in an RL problem:
 * <b>Training phase</b>: this is the time where the agent is exploring the diferent actions that could be taken in the diferent states and learning which is the best on each situation.
 * <b>Production phase</b>: this is the time where the agent applies the learned policy to get the best reward. In this phase, the policy is not changed.

The purpose of reinforcement learning is for the agent to learn the best policy that maximizes the obtained reward at the end of the episode. This means that the agent wants to take actions on the environment and get the maximum reward at the end which not alwais is to get the maximum reward for the next action.

In summary, <u>Reinforcement Learning is a machine learning paradigm that focuses on how an agent can learn to make optimal decisions in an uncertain environment through experience and feedback in the form of numerical rewards</u>. The environment is managed by a Markov decision process (given a particular state and action, the environment change to another state and gives a reward).

Understanding the basic components of Reinforcement Learning is essential for understanding and applying RL methods and algorithms. So let's go a bit deeper into this components:

## 2. Environment and Interpreter

As we have seen before, the environment component is our game board. It's where the actions are taken, the states are changed and the rewards are given.

In RL problems, while we are in the training phase, the environment component is normaly a simulator. This is because, as we have seen, RL agent needs to learn from the experience and that means that the agent needs simulate as much scenarios as posible in order to try to learn what happens on each state if it takes every posible actions, not just the next state, but also thelong-term expected return. For example, if it is playing a chess game, it could be good movement to sacrifice a pawn if it could give an advantage ten moves later.

So, as you can imagine, if it is not possible to play over the real environment a lot of times, it's important to have a good simulator of the real environment to be sure that the agent will learn good real policies.

Therefore, the use of simulations is an important aspect of RL and can significantly improve the performance of RL agents in real-world applications.

The components of a good simulator in Reinforcement Learning problems include:

 * <b>Accurate environment modeling</b>: The simulator should accurately model the dynamics of the environment, including the state transitions, actions, and rewards. <u>The simulator should be able to generate diverse and realistic scenarios</u> for the agent to learn from.
 * <b>High-quality data generation</b>: The simulator should be able to generate a large amount of data for the agent to learn from. The data should be diverse and representative of the real environment, while also being easy to preprocess and feed into the RL algorithm.
 * <b>Scalability and efficiency</b>: The simulator should be able to handle large-scale simulations and generate data quickly and efficiently. This is particularly important when training RL agents for complex environments or tasks.
 * <b>Flexibility and adaptability</b>: The simulator should be able to adapt to changes in the environment or task requirements, allowing the agent to continue learning and improving over time.
 * <b>Debugging and visualization tools</b>: The simulator should include debugging and visualization tools that allow developers to monitor and analyze the behavior of the agent during training. This can help identify and fix issues that may arise during the training process.
 
From the agent point of view, the simulator will be a black box where it receives the state, introduces an order (action to be taken) and get the reward and the next state.

It is normal to see the environment and the interpreter joined as one piece, since the interpreter work is to see the environment state and give the information to the agent, so, it could be seen as the environment output interface.

About the environment interface, as we have seen, there are two important components:

 * <b>Action space</b>: <u>This defines the set of actions that an agent can take in the environment</u>. The action space can be discrete or continuous, depending on the nature of the task. For example, in a game of chess, the action space is discrete, consisting of a finite set of possible moves, while in a self-driving car, the action space is continuous, consisting of a range of possible steering angles and velocities.

 * <b>State space</b>: <u>This defines the set of possible states that the environment can be in</u>. The state space includes all the relevant information that the agent needs to make decisions and take actions. For example, in a game of chess, the state space includes the positions of all the pieces on the board, while in a self-driving car, the state space includes information about the car's speed, position, and surrounding environment.

Together, the action space and state space define the interface between the agent and the environment in a Reinforcement Learning problem. The agent observes the current state of the environment and takes an action from the action space, and the environment responds with a new state and a reward signal that indicates how well the agent is performing.

## 3. Agent

As we have seen, the agent is the main component of interest in Reinforcement Learning. <u>The agent is responsible for making decisions based on the current state of the environment, taking actions, and receiving rewards from the environment. The ultimate goal of the agent is to learn an optimal policy that maximizes the cumulative reward over time.</u>

The agent can be implemented in various ways, including rule-based systems, decision trees, neural networks, or a combination of these approaches. The choice of implementation depends on the complexity of the task and the available resources.

The agent interacts with the environment by observing the current state, selecting an action from the available action space, and receiving a reward from the environment. The agent's decision-making process is guided by a policy, which maps states to actions. The policy can be deterministic or stochastic, depending on the nature of the task.

During the training phase, the agent learns from experience by updating its policy based on the rewards received and the resulting changes in the state of the environment. This process is known as the learning algorithm, and there are various methods available, including Q-learning, SARSA, and Deep Reinforcement Learning, among others.

In summary, the agent is the central component of RL, responsible for making decisions and learning from experience to maximize cumulative rewards over time.

### Define some constants for the learning process

As you have probably seen in previous lessons, a neural network training process needs some hiperparameters to be trained such as batch size, learning rate, etc.

Furthermore as we have seen before, the deep-Q learning algorithm need some more hiperparameters, like:
 * BUFFER_SIZE: it's the memory size or, in other words, how many experiences we are going to store to learning from.
 * GAMMA: this is the discount factor. This factor gives us the importance of the reward in subsequent steps with respect to the immediate reward of the next step. Its value will be between 0 and 1.
 * TAU: this parameter is used to update the target q-network softly. We need to have better prediction each time, but if it is changed radically, it can make not to converge the network.
 * UPDATE_EVERY: this is the parameter that configure how often the network is going to learn from the experiences.

```python
BUFFER_SIZE = int(1e5)  # replay buffer size
BATCH_SIZE = 64         # minibatch size
GAMMA = 0.99            # discount factor
TAU = 1e-3              # for soft update of target parameters
LR = 5e-4               # learning rate 
UPDATE_EVERY = 4        # how often to update the network

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")  # If we have a GPU, let's use it.
```

Ok, let's go with the agent implementation!!

In this cell, we are going to implement the Agent class with 5 methods:
 * <b>Constructor (\_\_init\_\_)</b>: this method initialice the local q-network, the target q-network, the optimizer (for local network learning) and the memory (see later the ReplayBuffer cell). At the end, the step counter is initialized to 0.
 * <b>step</b>: each step done in the environment is stored in the memory and, if it's time to learn, it's chosen a batch of experiences and they are use to learn from them (update the network to try to predict the future reward in a better way).
 * <b>act</b>: Returns actions for given state as per current policy. With a probability eps (the second argument), this method will give the best action (according to the network) to take or a random one. 
 * <b>learn</b>: Update value parameters using given batch of experience tuples. Local q-network learns from the experiences and target q-network is update softly from the local q-network parameters.
 * <b>soft_update</b>: Soft update model parameters for the target q-network.

```python
class Agent():
    """Interacts with and learns from the environment."""

    def __init__(self, state_size, action_size, seed):
        """Initialize an Agent object.
        
        Params
        ======
            state_size (int): dimension of each state
            action_size (int): dimension of each action
            seed (int): random seed
        """
        self.state_size = state_size
        self.action_size = action_size
        self.seed = random.seed(seed)

        # Q-Network
        self.qnetwork_local = QNetwork(state_size, action_size, seed).to(device)
        self.qnetwork_target = QNetwork(state_size, action_size, seed).to(device)
        self.optimizer = optim.Adam(self.qnetwork_local.parameters(), lr=LR)

        # Replay memory
        self.memory = ReplayBuffer(action_size, BUFFER_SIZE, BATCH_SIZE, seed)
        # Initialize time step (for updating every UPDATE_EVERY steps)
        self.t_step = 0
    
    def step(self, state, action, reward, next_state, done):
        # Save experience in replay memory
        self.memory.add(state, action, reward, next_state, done)
        
        # Learn every UPDATE_EVERY time steps.
        self.t_step = (self.t_step + 1) % UPDATE_EVERY
        if self.t_step == 0:
            # If enough samples are available in memory, get random subset and learn
            if len(self.memory) > BATCH_SIZE:
                experiences = self.memory.sample()
                self.learn(experiences, GAMMA)

    def act(self, state, eps=0.):
        """Returns actions for given state as per current policy.
        
        Params
        ======
            state (array_like): current state
            eps (float): epsilon, for epsilon-greedy action selection
        """
        # Epsilon-greedy action selection
        if random.random() > eps:
            state = torch.from_numpy(state).float().unsqueeze(0).to(device)
            self.qnetwork_local.eval()
            with torch.no_grad():
                action_values = self.qnetwork_local(state)
            self.qnetwork_local.train()

            return np.argmax(action_values.cpu().data.numpy())
        else:
            return random.choice(np.arange(self.action_size))

    def learn(self, experiences, gamma):
        """Update value parameters using given batch of experience tuples.

        Params
        ======
            experiences (Tuple[torch.Variable]): tuple of (s, a, r, s', done) tuples 
            gamma (float): discount factor
        """
        # Obtain random minibatch of tuples from D
        states, actions, rewards, next_states, dones = experiences

        ## Compute and minimize the loss
        ### Extract next maximum estimated value from target network
        q_targets_next = self.qnetwork_target(next_states).detach().max(1)[0].unsqueeze(1)
        ### Calculate target value from bellman equation
        q_targets = rewards + gamma * q_targets_next * (1 - dones)
        ### Calculate expected value from local network
        q_expected = self.qnetwork_local(states).gather(1, actions)
        
        ### Loss calculation (we used Mean squared error)
        loss = F.mse_loss(q_expected, q_targets)
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

        # ------------------- update target network ------------------- #
        self.soft_update(self.qnetwork_local, self.qnetwork_target, TAU)                     

    def soft_update(self, local_model, target_model, tau):
        """Soft update model parameters.
        _target = *_local + (1 - )*_target

        Params
        ======
            local_model (PyTorch model): weights will be copied from
            target_model (PyTorch model): weights will be copied to
            tau (float): interpolation parameter 
        """
        for target_param, local_param in zip(target_model.parameters(), local_model.parameters()):
            target_param.data.copy_(tau*local_param.data + (1.0-tau)*target_param.data)
```

OK, now that we have the Agent class, let's do the memory class to store the experiences and get samples for training.

For the memory we will use the "deque" library. The "deque" library in Python is a data structure that represents a double-ended queue, which is a list-like container that supports adding and removing elements from both ends efficiently. One of the key features of "deque" is the ability to specify a maximum length using the "maxlen" parameter. When a "deque" object is initialized with a maximum length, it becomes a fixed-size container that discards the oldest elements when new elements are added and the maximum length is exceeded. This makes "deque" a useful tool for implementing memory buffers with limited capacity, where older data can be automatically purged to make room for newer data.

Another important point in our code is the way a sample set of experiences is selected. We can use random.sample method to select a ramdom set of experiences from the deque object.

So le's will implement the memory (ReplayBuffer) with the methods:
 * <b>Constructor (\_\_init\_\_)</b>: to initialize all the needed structures.
 * <b>add</b>: to add experiences to the memory.
 * <b>sample</b>: to get a set of experiences of self.batch_size size.
 * <b>\_\len\_\_</b>: to know how many experiences there are stored.

```python
class ReplayBuffer:
    """Fixed-size buffer to store experience tuples."""

    def __init__(self, action_size, buffer_size, batch_size, seed):
        """Initialize a ReplayBuffer object.

        Params
        ======
            action_size (int): dimension of each action
            buffer_size (int): maximum size of buffer
            batch_size (int): size of each training batch
            seed (int): random seed
        """
        self.action_size = action_size
        self.memory = deque(maxlen=buffer_size)  
        self.batch_size = batch_size
        self.experience = namedtuple("Experience", field_names=["state", "action", "reward", "next_state", "done"])
        self.seed = random.seed(seed)
    
    def add(self, state, action, reward, next_state, done):
        """Add a new experience to memory."""
        e = self.experience(state, action, reward, next_state, done)
        self.memory.append(e)
    
    def sample(self):
        """Randomly sample a batch of experiences from memory."""
        experiences = random.sample(self.memory, k=self.batch_size)

        states = torch.from_numpy(np.vstack([e.state for e in experiences if e is not None])).float().to(device)
        actions = torch.from_numpy(np.vstack([e.action for e in experiences if e is not None])).long().to(device)
        rewards = torch.from_numpy(np.vstack([e.reward for e in experiences if e is not None])).float().to(device)
        next_states = torch.from_numpy(np.vstack([e.next_state for e in experiences if e is not None])).float().to(device)
        dones = torch.from_numpy(np.vstack([e.done for e in experiences if e is not None]).astype(np.uint8)).float().to(device)
  
        return (states, actions, rewards, next_states, dones)

    def __len__(self):
        """Return the current size of internal memory."""
        return len(self.memory)
```
